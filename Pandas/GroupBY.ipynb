```python
genres.groups
genres.describe()
```

```python
#agg method 
#passing dict
genres.agg(
    {
        'Runtime':'min',
        'Gross':'sum',
        'IMDB_Rating':'mean',
        'No_of_Votes':'max'
    }
)
```

```python
#split (apply) combine 
#apply --> builtin functions 
genres.apply(min).head(5)
```

```python
#find number of movies starting with A for Each Group 

def foo(group):
    return (group['Series_Title'].str.startswith('A')).sum()
    
```

```python
genres.apply(foo)
```

```python
#find ranking of each movies in the group according to IMDB score
def rank_movie(group):
    group['genre_rank'] = group['IMDB_Rating'].rank(ascending=False)
    return group
```

```python
genres.apply(rank_movie).sample(5)
```

```python
#find the normalize IMDB rating groups wise 
def normal(group):
    group['normalized_rating'] = (group['IMDB_Rating'] - group['IMDB_Rating'].min())/(group['IMDB_Rating'].max() - group['IMDB_Rating'].min())
    return group
```

```python
genres.apply(normal).sample(6)
```

```python
# groupby on multiple cols

duo = movies.groupby(['Director', 'Star1'])

# size
duo.size()

# get_group
duo.get_group(('Aamir Khan', 'Amole Gupte'))
```

```python
#find the best (in - terms of metascore(avg) actors --> genre combo)
movies.groupby(['Star1', 'Genre'])['Metascore'].mean().reset_index().sort_values('Metascore', ascending=False)
```
