Series Attribute
----------------
The `Series` attribute is a powerful tool in data visualization that allows you to group and differentiate data points based on specific categories or characteristics. It is commonly used in various types of charts and graphs, such as line charts, bar charts, and scatter plots, to visually distinguish between different sets of data.


```python
#coustum Index
 
import pandas as pd
import numpy as np

marks = [67,78,88,100,89]
subject =['maths','English','marathi','Dsa','SC']

marks = pd.Series(marks,index=subject,name = 'Rushi che marks')
marks
```

```python
#size
marks.size
```

```python
#dtype
marks.dtype
```

```python
#name 
marks.name
```

```python
#is_unique

marks.is_unique
pd.Series([11,22,33,44,44]).is_unique
```

```python
#index
marks.index
```

```python
#value
type(marks.values)
marks.values
```
