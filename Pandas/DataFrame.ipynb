```python
#find all the final winner
mask = ipl['MatchNumber'] == 'Final'
new_df = ipl[mask]
new_df[['Season','WinningTeam','Venue']]

ipl[ipl['MatchNumber'] == 'Final'][['Season','WinningTeam']]
```

```python
#how many super overs finisher have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
#how many matches has RCB won in kolkata

ipl[(ipl['City'] == 'Bangalore') & (ipl['WinningTeam'] == 'Royal Challengers Bangalore')].shape[0]
```

```python
#Toss Winner is Match winner in percentage
(ipl[ipl['TossWinner'] == ipl['WinningTeam']].shape[0]/ipl.shape[0])*100
```

```python
#movies with rating higher than 8 and voting > 1000
movies[(movies['imdb_rating'] > 8.5) & (movies['imdb_votes'] > 1000)].shape[0]
```

```python
#Action Movies with rating higher than 7.5
mask1 = movies['genres'].str.split('|').apply(lambda x : 'Action' in x)
mask2 = movies['imdb_rating'] > 7.5

movies[mask1 & mask2].shape[0]
```
