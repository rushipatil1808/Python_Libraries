Importing Pandas 

What is Series in Pandas
--> A pandas series is like a column in a table it is 1-D array holding of any type

```python
import pandas as pd
import numpy as np
```

```python
#strings
country = ['India','Pakistan','USA','Nepal','China']
pd.Series(country)
```

```python
#Integer 
runs = [33,45,66,77,22,55]
pd.Series(runs)
```

```python
#coustum Index

marks = [67,78,88,100,89]
subject =['maths','English','marathi','Dsa','SC']

pd.Series(marks,index=subject)
```

```python
#setting a name 
marks = pd.Series(marks,index=subject,name = 'Rushi che marks')
marks
```

```python
marks = {
    'maths':67,
    'English':78,
    'marathi':88,
    'Dsa':100,
    'SC':89
}

marks_series = pd.Series(marks,name = 'Rushi che marks')
marks_series
```
