```python
import numpy as np
import pandas as pd
batsman = pd.read_csv(r'E:\PythonLibraries\CSV\sharma-kohli.csv')
batsman

plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted')
plt.plot(batsman['index'],batsman['RG Sharma'],color='black',linestyle='dashed')

plt.title('Rohit Sharma vs Virat Kohli career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Score')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'],
         color='#09E10E', linestyle='solid', linewidth=3, marker='D',label='Virat')

plt.plot(batsman['index'], batsman['RG Sharma'],
         color='#FC00D6', linestyle='dashdot', linewidth=2, marker='D',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend()
plt.grid()
```

Scatter Plot-->
A scatter plot is a type of data visualization that displays the relationship between two numerical variables. Each point on the plot represents an observation in the dataset, with its position determined by the values of the two variables being compared.

```python
#plt.scatter simple function

x = np.linspace(-10,10,50)

y = np.random.randn(50)
plt.scatter(x, y)
```

```python
#plt_scatter on pandas data 

batter = pd.read_csv(r'E:\PythonLibraries\CSV\batter.csv')
batter.head(50)
```

```python

plt.scatter(batter['avg'],batter['strike_rate'],color='red',marker='+')
plt.title('Avg and Sr analysis of to 50')
plt.xlabel('Average')
plt.ylabel('SR')
```

Bar Chart 


```python
#simple bar chart 
ch = [10,50,30,40,60]
colors = ['red','blue','green','yellow','pink']

plt.bar(colors,ch,color = 'Red')
```

```python
df = pd.read_csv(r'E:\PythonLibraries\CSV\batsman_season_record.csv')
df 
```

```python
plt.bar(np.arange(df.shape[0]) - 0.2, df['2015'],width=0.2)
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2)
plt.bar(np.arange(df.shape[0]) + 0.2, df['2017'],width=0.2)

plt.xticks(np.arange(df.shape[0]), df['batsman'],rotation = 90)
plt.show()



#plt.bar(df['batsman'],df['2016'])
#plt.bar(df['batsman'],df['2017'])
```

```python
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016'] + df['2017'],label='2015')

plt.legend()
plt.show()
```

```python
#data Set
gayle = pd.read_csv(r'E:\PythonLibraries\CSV\gayle-175.csv')
gayle
```

```python
plt.pie(gayle['batsman_runs'],labels=gayle['batsman'],autopct='%0.1f%%')

plt.show()
```

```python
explode = [0] * len(gayle)
explode[-1] = 0.2

plt.pie(
    gayle['batsman_runs'],
    labels=gayle['batsman'],
    autopct='%0.1f%%',
    explode=explode,
    shadow=True
)
plt.show()
```
