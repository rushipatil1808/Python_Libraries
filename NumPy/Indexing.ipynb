Indexing And Slicing

```python
import numpy as np

a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(8).reshape(2,2,2) 

a1
```

```python
a2
```

```python
a2[1,0]
```

```python
a3
```

```python
a3[1,0,1]
```

```python
a3[0,1,0]
```

```python
a3[0,0,0]
```

```python
a3[1,1,0]
```

Slicing

```python
a1
```

```python
a1[2:5]
```

```python
a2
```

```python
a2[0,:]
```

```python
a2[:,2]
```

```python
a2[1:,1:3]
```

```python
a2[::2,::3]
```

```python
a2[::2,1::2]
```

```python
a2[1,::3]
```

```python
a2[0:2,1:]
```

```python
a3 = np.arange(27).reshape(3,3,3)
a3
```

```python
a3[0,1,:]
```

```python
a3[1,:,1]
```

```python
a3[2,1:,1:]
```

```python
a3[::2,0,::2]
```
