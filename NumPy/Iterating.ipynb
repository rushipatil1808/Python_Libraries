Iterating

```python
import numpy as np

a1 = np.arange(10)
a2 = np.arange(12).reshape(3,4)
a3 = np.arange(27).reshape(3,3,3)
```

```python
a1

for i in a1:
    print(i)
```

```python
a2

for i in a2:
    print(i)
```

```python
a3

for i in a3:
    print(i)
```

```python
for i in np.nditer(a3):
    print(i)
```

```python
#reshape
a4 = a1.reshape(2,5)
a4
```

```python
#transppse
np.transpose(a2)
```

```python
#ravel
a3.ravel()
```

Stackig

```python
#horizontal Staking 

a4 = np.arange(12).reshape(3,4)
a5 = np.arange(12,24).reshape(3,4)
a5
```

```python
np.hstack((a4,a5))
```

```python
#vertical Staking
#its use in multiple data set to analise the data then to store like this sevral of data 
np.vstack((a4,a5))
```

```python
#horizontal Spliting 

np.hsplit(a5,2)
```

```python
#vertical spliting

np.vsplit(a5,3)
```
