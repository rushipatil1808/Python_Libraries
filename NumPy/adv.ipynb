```python
# speed

# list can store non homogenios data 

a = [i for i in range(1000000)]
b = [i for i in range(1000000,2000000)]

c = []

import time

start = time.time()

for i in range(len(a)):
    c.append(a[i] + b[i])

print(time.time()-start)
```

```python
#numpy 

import numpy as np

a = np.arange(1000000)
b = np.arange(1000000,2000000)

start = time.time()
c = a + b

print(time.time()-start)
```
