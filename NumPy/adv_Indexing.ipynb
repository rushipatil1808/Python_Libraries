```python
#Normal indexing and Slicing 

import numpy as np

a = np.arange(24).reshape(6,4)
a
```

```python
a[1,2]
```

```python
a[1:3,1:3]
```

```python
#Fancy Indexing 

a[[0,2,3]]
```

```python
a[:,[0,2,3]]
```

```python
# Boolean Indexing

import numpy as np
a = np.random.randint(1,100,24).reshape(6,4)
a
```

```python
# find all numbers greater than 50

#a > 50
a[a > 50]
```

```python
#find out a evan numbers 
a[a  % 2 == 0 ]
```

```python
#find all number greater than 50 in even 

a [(a > 50) & (a % 2 == 0)]
```

```python
#find all numbers not divisible by 7
a[~(a % 7 == 0)]
```
