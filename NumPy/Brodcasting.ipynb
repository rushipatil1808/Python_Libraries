broadcasting's
=================
Broadcasting is a powerful mechanism that allows NumPy to perform operations on arrays of different shapes and sizes. 

It enables you to apply operations to arrays without explicitly reshaping them, making your code more concise and efficient.

```python
#same shape
import numpy as np

a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)


print(a)
print(b)

print(a  + b)
```

```python
import numpy as np

a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a + b)
```

Broadcasting Rules.

Rule 1: Same Number of Dimensions

If two arrays have different dimensions, NumPy adds 1 to the left side of the smaller array’s shape.

2. Make each dimension of the two arrays the same size.
   • If the sizes of each dimension of the two arrays do not match, dimensions with size 1 are stretched to the size of the other array.
   
   • If there is a dimension whose size is not 1 in either of the two arrays, it cannot be broadcasted, and an error is raised.

```python
import numpy as np
a = np.array([1, 2, 3])      # shape = (3,)
b = np.array([[1], [2], [3]]) # shape = (3,1)


print(a)
print(b)
print(a + b)
```

```python
a = np.arange(3).reshape(1,3)
b = np.arange(3).reshape(3,1)

print(a)
print(b)

print(a+b)
```
