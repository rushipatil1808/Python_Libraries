broadcasting's
=================
Broadcasting is a powerful mechanism that allows NumPy to perform operations on arrays of different shapes and sizes. 

It enables you to apply operations to arrays without explicitly reshaping them, making your code more concise and efficient.

```python
#same shape
import numpy as np

a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)


print(a)
print(b)

print(a  + b)
```

```python
import numpy as np

a = np.arange(6).reshape(2,3)
b = np.arange(3).reshape(1,3)

print(a)
print(b)

print(a + b)
```
